## Removing columns having more than 40% of null values

```python
#Drop the columns which are having more than 40% of the values are having null values.
#function that will remove the cols having null values based on cut off.
def drop_null_vals(data_set, cut_off):
    empty_vals = list(data_set.columns[100*data_set.isnull().mean() > cut_off])
    print(f"List of columns having more than {cut_off}% empty values: {empty_vals}\n")
    data_set = data_set.drop(empty_vals, axis=1)
    return data_set

curated_data = drop_null_vals(input_data, 40)
```

### Imputing the numerical columns with median and categorical columns with mode.

- Using the median is a reliable approach to impute numerical data as it does not exaggerate the data and if there is any outliers then median is the safe possible approach.
- In this case, we are using the mode to impute the categorical data.

```python
curated_data[categorical_cols].isnull().sum()
```

```python
#Function that will impute the missing values with mode in categorical columns.
def impute_mod_val(data_set, list_of_cols):
    for col in list_of_cols:
        data_set[col] = data_set[col].fillna(data_set[col].mode()[0])
    return data_set
```

```python
imputed_data = impute_mod_val(curated_data, categorical_cols)
```

```python
imputed_data[categorical_cols].isnull().sum()
```

```python
#Function that will impute the missing values with median in numerical columns.
def impute_med_val(data_set, list_of_cols):
    for col in list_of_cols:
        data_set[col] = data_set[col].fillna(data_set[col].median())
    return data_set
```

```python
imputed_data = impute_med_val(curated_data, numerical_cols)
```

## Key Takeaways

- Age group : 
    - It is safer to grant loans for 30 to 40 yr old clients
- Employement Group :
    - Prioritize loans for those with 5-10 years of employment to lower default rates.
    - It is reccomended to grant loan to group of people with more work experience.
- Occupation Type : 
    - It is recommended to grant loans to people with occupations as labourers and sales staff.
- Education Qualification : 
    - It is recommended to grant loans to individuals who have completed at least a higher education level or higher.
- Income group:
    - It is recommended to grant loans to individuals who have income more than 80k.
    - It is risky to provide loan to the individual with low income
- Housing Type:
    - It is recommended to grant loans to individuals who statys with parents.
    - It is risky to provide loan to the individual who own a house or rented apartment.
- Bivariate Analysis :
    - It is safer to grant loan to the individual having high correlation between amount annuity and amount of good price as they are more likely to be able to repay the loan.
    - A strong correlation between AMT_CREDIT and AMT_GOODS_PRICE suggests clients often borrow an amount close to the price of the goods they intend to purchase.
- Loan Contract Status: 
    - Loans which were canceled or refuse, indicates that clients did not match the expectations and it is riskier to provide loan to such clients.
- Client Type Status:
    - It is benificial to grant loan to the client whose loan got approved previously as well.
